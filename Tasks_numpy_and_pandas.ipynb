# Задача 1 (numpy).
Сгенерировать квадратную матрицу случайных чисел (произвольного размера). Написать
алгоритм поиска седловых точек. В качестве седловых точек рассматриваются значения
массива, которые являются максимальными в строке и одновременно минимальными в
столбце. Алгоритм должен возвращать координаты седловых точек (индексы по строке и
столбцу). В случае, если седловые точки отсутствуют, алгоритм должен вернуть пустой
список/картеж. Запрещается использовать циклы, запрещается использовать библиотеку
pandas.

```python
n = 10
matrix = np.random.randint(0, 100, (n, n))
# matrix = np.array([[1,0,1],[2,2,0],[3,2,0],[2,2,0],[2,2,0],[2,2,0],[2,2,0],[2,2,0],[2,2,0],[2,2,0],[2,2,0],[2,2,0],[1,0,1]])
matrix
```

```python
max_in_line = np.amax(matrix, axis=1)
min_in_column = np.amin(matrix, axis=0)
common_elements = np.intersect1d(max_in_line, min_in_column)
common_elements
```

```python
line_indices = np.where(np.isin(max_in_line, common_elements))[0]
column_indices = np.where(np.isin(min_in_column, common_elements))[0]
line, column = np.meshgrid(line_indices, column_indices)

saddles = np.stack((line.ravel(), column.ravel()), axis=-1)
print(saddles)
```

# Задача 2 (pandas).
Сгенерировать DataFrame c колонками [user_id, item_id, time], заполненный случайными
целыми числами. DataFrame должен иметь 1.000 строк. Элементы в столбце user_id
должны принимать значения в интервале [0,10]. Элементы в колонках item_id и time могут
принимать произвольные целочисленные значения. Будем считать, что полученный
DataFrame есть некоторый набор покупок пользователей. То есть некоторый пользователь
user_id купил товар item_id в момент времени time. Необходимо написать алгоритм,
который для каждого пользователя в DataFrame сохранит последние 5 покупок. 

```python
n=1000
df = pd.DataFrame({
    'user_id': np.random.randint(0, 11, n), 
    'item_id': np.random.randint(0, n, n),
    'time': np.random.randint(1, n, n)})
df
```

```python
df = df.sort_values('time', ascending=False)
df = df.groupby('user_id').head(5)
df
```
